```python
def countinv(arr, l, r, counts):
    if l < r:
        m = (r + l) // 2
        
        countinv(arr, l, m, counts)
        countinv(arr, m + 1, r, counts)
        countandmerge(arr, l, m, r, counts)

def countandmerge(arr, l, m, r, counts):
    n1 = m - l + 1
    n2 = r - m
    left = arr[l:m + 1]
    right = arr[m + 1:r + 1]
    left_indices = list(range(l, m + 1))
    right_indices = list(range(m + 1, r + 1))
    
    i = 0
    j = 0
    k = l
    while i < n1 and j < n2:
        if left[i] <= right[j]:
            arr[k] = left[i]
            counts[left_indices[i]] += j  # Add the count of elements already merged from the right
            i += 1
        else:
            arr[k] = right[j]
            j += 1
        k += 1
    while i < n1:
        arr[k] = left[i]
        counts[left_indices[i]] += j
        i += 1
        k += 1
    while j < n2:
        arr[k] = right[j]
        j += 1
        k += 1

class Solution(object):
    def countSmaller(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        n = len(nums)
        counts = [0] * n
        countinv(nums, 0, n - 1, counts)
        return counts
    
if __name__ == "__main__":
    nums = [2, 0, 1]
    solution = Solution()
    print(solution.countSmaller(nums))  # Expected output: [2, 1, 0]
```

```python
def countinv(indices, l, r, counts, nums):
    if l < r:
        m = (r + l) // 2
        
        countinv(indices, l, m, counts, nums)
        countinv(indices, m + 1, r, counts, nums)
        countandmerge(indices, l, m, r, counts, nums)

def countandmerge(indices, l, m, r, counts, nums):
    n1 = m - l + 1
    n2 = r - m
    left = indices[l:m + 1]
    right = indices[m + 1:r + 1]
    
    i = 0
    j = 0
    k = l
    while i < n1 and j < n2:
        if nums[left[i]] <= nums[right[j]]:
            indices[k] = left[i]
            counts[left[i]] += j  # Add the count of elements already merged from the right
            i += 1
        else:
            indices[k] = right[j]
            j += 1
        k += 1
    while i < n1:
        indices[k] = left[i]
        counts[left[i]] += j  # Add the count of elements already merged from the right
        i += 1
        k += 1
    while j < n2:
        indices[k] = right[j]
        j += 1
        k += 1

class Solution(object):
    def countSmaller(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        n = len(nums)
        counts = [0] * n
        indices = list(range(n))  # Track the original indices
        countinv(indices, 0, n - 1, counts, nums)
        return counts
    

nums = [2, 0, 1]
solution = Solution()
print(solution.countSmaller(nums))  # Expected output: [2, 1, 0]
```
